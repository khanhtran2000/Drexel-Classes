### Measuring The Effectiveness of A Good Prediction/Clustering Algorithm 

A good prediction/clustering algorithm starts with a good dataset. Well actually, it starts with two good datasets, one train set and one test set. Multiple tests can be carried out the same way. Cross validation is a more sophisticated approach that can also be used which can create quick, unbiased, and effective performance tests. 

Another approach is testing algorithms. Here, these algorithms do not include our self-built ones, but are the other available algorithms that can ne used to solve the same problem. For example, both KNN and logistic regression can in some cases solve the same classification problem, so it's helpful to examine the performance of different models. #### Outliers

For detecting outliers, we will examine the following columns: 
- Severity
- Distance(mi)
- Temperature(F)
- Wind_Chill(F)
- Humidity(%)
- Pressure(in)
- Visibility(mi)
- Wind_Speed(mph)
- Precipitation(in)

These attributes are statistics measurements, so it is reasonable to examine them for potential outliers. The approach in used here is to calculate z-score based on mean and standard deviation of each sample.

```python
columns = [
    "Severity", "Distance(mi)", "Temperature(F)", "Wind_Chill(F)",
    "Humidity(%)", "Pressure(in)", "Visibility(mi)", "Wind_Speed(mph)",
    "Precipitation(in)"
    ]

#  Using z_score
def detect_outlier(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for value in data:
        z_score = (value-mean)/std
        
        if np.abs(z_score) > threshold:
            outliers.append(value)
    
    return outliers

# Iterate through each attribute in the list
for column in columns:
    outliers = detect_outlier(df[column].values)
    print("The outliers in column {}:".format(column))
    print(outliers)
    print("Total number:", len(outliers))
    
```

According to the result, "Severity" and "Distance(mi)" are the only two columns that possess outliers. However, the outliers in each of these attributes are not mistakes. 

For "Severity", the outliers all have the value of 4. This is an error since the severity scale is from 1 to 4 with 4 indicating the heaviest impact. The reason why they are detected as outliers is because there are only 112,320 cases where the severity level is 4, and it is an insignificant number compared to the total of over 3.5 million cases. 

For "Distance(mi)", we see the need for further examination. We plot a normal distribution of the outliers sample of this attribute in order to detect the data points that significantly outstand the rest. 

```python
outliers_distance = detect_outlier(df["Distance(mi)"].values)

# Min of the "Distance(mi)" outliers 
min(outliers_distance)
```

```python
# Mean of the outliers from "Distance(mi)"
mean = np.mean(outliers_distance)
print(mean)

# Standard deviation of the outliers from "Distance(mi)"
std = np.std(outliers_distance)
print(std)

# Normal distribution of the outliers
outliers_distance.sort()
x = []

for outlier in outliers_distance:
    z_score = (outlier-mean)/std
    x.append(z_score)

plt.plot(x, stats.norm.pdf(x,0,1))
plt.show()
```

```python
# Outliers in the "Distance(mi)" outliers sample
n = 0
m = 0 

for i, z_score in enumerate(x): 
    if z_score > 3:
        print("Outlier:", outliers_distance[i])
        print("z score:", z_score)
        n += 1
        
        if outliers_distance[i] > 100: m += 1

print("\n")
print("Number of highly-impacted outliers:", n)
print("Number of outliers over 100 miles:", m)
```

There are 37,757 outliers for this attribute, meaning that there is only 1.07% of the total cases where the affected length of road extended to over 4.933 miles. This is rather a small percentage, but after further investigating, we notice that the normal distribution graph is highly skewed to the right, which indicates a small number of outstanding outliers. We print out the outliers that have z-score higher than 3, and as expected, there are 41 cases where the affected length of road extended to over 100 with an exceptional case of 333.63 miles. The rest range from 38 to 99 miles, and the total 543 cases of "highly-impacted outliers".

### Quick Statistics

This section will focus on providing simple statistics for the most important attributes and describe what they mean. First of all, we will look at "Severity", "Start_Time","State", and "Temperature(F)".

```python
# Severity

# Mean
mean_sev = np.mean(df["Severity"].values)
# Mode
mode_sev = stats.mode(df["Severity"].values)[0]
# Count for mode
count_mode_sev = stats.mode(df["Severity"].values)[1]

print("Mean of 'Severity':", mean_sev)
print("Mode of 'Severity':", mode_sev[0])
print("Count of Mode:", count_mode_sev[0])
```

On average, the severity level is 2.34. Considering 1 is the least impact on traffic and 4 is a significant one, 2.34 is slightly above the average level of impact. There are 2,373,210 cases that fall into level 2 of severity, which is also the mode of this attribute. In other words, based on the dataset, every 2 out of 3 accidents cause a severity of level 2. 

Next, let's look at "Start_Time". However, we won't go into details here. We only look at the starting hours of the accident to find out at what time of a day did the accidents happen the most.

```python
# Start_Time

# Convert to datetimes 
df["Start_Time"] = pd.to_datetime(df["Start_Time"])

# Extract starting hours
start_hours = pd.DatetimeIndex(df["Start_Time"]).hour

# Mode of starting hours
mode_start = stats.mode(start_hours)[0]
# Counts for mode
counts_mode_start = stats.mode(start_hours)[1]

print("The time when accidents happened the most is:", mode_start[0])
print("Number of such cases:", counts_mode_start[0])
```

The numbers speak for themselves here. We find out that the largest number of cases happened in the morning at around 8am. Logically, this also fits the fact that 8am is usually one of the rush hours of the day when the roads are the most crowded. 

Lastly, we will look at the temperatures during the accidents. Since column "Temperature(F)" has a lot of missing values, we will leave them out before conducting calculations.

```python
# Temperature(F)

# Leave out missing values
temps = [temp for temp in df["Temperature(F)"].values if np.isnan(temp)==False]
# Mean
mean_temp = np.mean(temps)
# Mode
mode_temp = stats.mode(temps)[0]
# Count for mode+-2 Farenheit degreee
count_mode_temp = stats.mode(temps)[1] + temps.count(mode_temp[0]-2) + temps.count(mode_temp[0]+2)

print("Mean of 'Temperature(F)':", mean_temp)
print("Mode of 'Temperature(F)':", mode_temp[0])
print("Count of Mode+-2 Farenheit degreee:", count_mode_temp[0])
```

According to the result, we know that the average temperature at the time of the accidents is 61.94 F. The temperature that appear the most time is 68.0 F. With temperature, we have a more flexible counting approach as this attribute has a rather high variance. We increase the range to 66.0 to 70.0 F and are able to find out over 200,000 cases fall into this range of temperature.

First of all, let's try to visualize the relationship between "Weather_Condition" and the (starting) months of each accidents.

```python
# Explore unique weather conditions
unique_cons = df["Weather_Condition"].unique()
unique_cons
```

"Weather_Condition" contains a long list of unique descriptions. On one hand, this is very informative which will support the precision of our predicting model. On the other hand, this raise a difficulty for us as we want to simplify the features for the sake of general visualization and data understanding. Therefore, we will simplify this class into two categories only: rain and no rain. 

```python
# Create a new dataframe with weather conditions cases counts. This automatically drops the nan values
count_weather = df.groupby(["Weather_Condition"]).size().reset_index(name="Cases_Count")

# Rain or not rain
rain_cons = []
for con in count_weather["Weather_Condition"].values:
    if "Rain" in con.split(" "):
        rain_cons.append("Rain")
    else:
        rain_cons.append("Not rain")

# New column
count_weather['Rain or not'] = rain_cons

# New dataframe 
count_rain = count_weather.groupby(["Rain or not"]).Cases_Count.sum().reset_index(name="Cases Count")

print(count_rain)

# Plotting
fig = plt.figure()
ax = count_rain.plot.bar("Rain or not", "Cases Count", color='magenta', figsize=(12,7))
ax.set_title("Cases Count when Rain and Not Rain", size=14)
ax.set_ylabel("Cases Count", size=12)
plt.show()
```

So this quick visualization has shown us that accidents tend to happen when it's not raining at least in the scope of this dataset. There is only ~252,000 cases occured in the rain. However, this is just an outcome from a simplified problem. The real question is to compare cases count of each unique weather conditions. It will be discussed in further length in our final report with the additional analysis on "Precipitation" attribute.

Now, let's look at how months and weather conditions correlate. This time, we will categorize the weather into five classes: "Thunderstorm", "Rain", "Snow", "Fog", and "Others".

```python
# Extract starting months
start_months = pd.DatetimeIndex(df["Start_Time"]).month

# New dataframe of weather and months information
weather_months = pd.DataFrame({"Start_Month":start_months, "Weather_Condition":df["Weather_Condition"].values})

# Drop rows with nan values of weather condition
weather_months = weather_months[weather_months["Weather_Condition"].isin(count_weather["Weather_Condition"].values)]

# Rewrite the weather conditions 
new_cons = []
for con in weather_months["Weather_Condition"].values:
    if "Thunderstorm" in con.split(" ") or "Thunderstorms" in con.split(" ") or "T-Storm" in con.split(" "):
        new_cons.append("Thunderstorm")
    elif "Rain" in con.split(" "):
        new_cons.append("Rain")
    elif "Snow" in con.split(" "):
        new_cons.append("Snow")
    elif "Fog" in con.split(" "):
        new_cons.append("Fog")
    else:
        new_cons.append("Others")

# New column
weather_months["Weather_Condition"] = new_cons

weather_months = weather_months.groupby(["Start_Month", "Weather_Condition"]).size().reset_index(name="Cases Count")

# Create a pivot
pivot_weather_months = weather_months.pivot(index='Start_Month', columns='Weather_Condition', values='Cases Count')
pivot_weather_months
```

```python
# Plotting
colors = ['#054B03','#056402','#067D02','#099E04','#50EC01']

# Note: We are rearranging the layer ordering 
pivot_weather_months.loc[:,['Others','Fog','Snow','Rain','Thunderstorm']].plot.bar(stacked=True, color=colors, figsize=(12,7))
plt.title("Cases Counts based on Months and Simplified Weather Conditions", size=14)
plt.ylabel("Cases Count", size=12)
plt.xlabel("Start Month", size=12)
```

```python
# Leave out the Others
pivot_weather_months.loc[:,['Fog','Snow','Rain','Thunderstorm']].plot.bar(stacked=True, color=colors[1:], figsize=(12,7))
plt.title("Cases Counts based on Months and Simplified Weather Conditions (No Others)", size=14)
plt.ylabel("Cases Count", size=12)
plt.xlabel("Start Month", size=12)
```

According to these two graphs, we can say that most of the accidents did not happen on days with thunderstorm, rain, snow, and fog but they happened in days with other weather condition. Perhaps, in reality, normal days happen much more often that days with such extreme conditions. However, this will need further investigation. An interesting pattern that exists in both the graphs is the tendency of less cases happening in extreme weathers during the summer. This may be a result of in reality, we tend to have more "nice weather" in this season. The total number of cases in July is also the lowest, indicating extreme weathers may cause more car accidents. If we leave out the other conditions and just look at the first four, "Rain" seems to appear the most frequently during accidents. "Snow" does occur during the start and end of a year, and "Thunderstorm" happens most frequently during summer.

## Exceptional Work

We have explored the "Start_Time" attribute, but we didn't pay much attention to its partner, the "End_Time". Here, let's look at the "End_Time", but not to plot out another bar chart like what we did with the "Start_Time". This time, we will explore the durations of the accidents. 

```python
# End_Time

# Convert to datetimes 
df["End_Time"] = pd.to_datetime(df["End_Time"])

# Calculate the duration of accidents in hours
time_difference = (df["End_Time"] - df["Start_Time"])
durations = [diff.total_seconds()/3600 for diff in time_difference]

# Mean duration
mean_dur = np.mean(durations)

# Mode duration
mode_dur = stats.mode(durations)[0]
# Counts for mode
counts_mode_end = stats.mode(durations)[1]

print("The mean duration of the accidents is {} hour(s).".format(mean_dur))
print("Most of the accidents lasted about {} hour(s).".format(mode_dur[0]))
print("Number of such cases:", counts_mode_end[0])
print("The longest duration is {} hour(s).".format(max(durations)))
```

What's surprising here is not the mean, not the mode, but the longest duration of them all. While most of the accidents lasted about 6.0 hours, and the average is just under 2.0 hours which means people usually need much less than 10 hours to clean up everything, this one particular accident needs 23,699 hours to do so. That is 987 days, or nearly 3 years! This leads us to further investigate the case.
