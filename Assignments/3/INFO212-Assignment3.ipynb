Wikipedia has a Webpage: [List of countries by past and future population](https://en.wikipedia.org/wiki/List_of_countries_by_past_and_future_population). On this page, there are three tables containing the estimated populations from 1950 to 2050 for all the countries. Read the description on the Webpage to understand the information in the tables. Complete the steps below.
- Download the Webpage as an HTML file in your disk. 
- Read in the tables using Pandas' `read_html` method. Answer the following question:
 * How many tables have you read in?
- Show the first 5 rows of the DataFrame correpsonding to the table "Estimates between the years 1950 and 1980 (in thousands)"
- How many countries does the Wikipedia page list their populations?
- Create a DataFrame containing all the data from the population tables. I.e., the DataFrame contains the popultation and average annual growth data from 1950-2050. You need to combine the three population tables.

![](https://i.imgur.com/ImXjIo0.png)

```python
import numpy as np
import pandas as pd
np.random.seed(12345)
import matplotlib.pyplot as plt
#plt.rc('figure', figsize=(10, 6)) # This doesn't work on my Jupyter notebook
np.set_printoptions(precision=4, suppress=True)
%matplotlib inline

# Display all columns
pd.set_option('display.max_columns', None)  

# Using read_html() to read the tables 
# The link to the html file will be different on other devices
tables = pd.read_html("/Users/macbook/Desktop/INFO212/Assignment 3/country_population.html")

#---------------------------------------------------------------------------
# How many tables have you read in?


# Using len() to get the number of tables that were read in
print("* I got", len(tables), 'tables.')
print("\n")

#---------------------------------------------------------------------------
# Show the first 5 rows of the DataFrame correpsonding to the table "Estimates between the 
# years 1950 and 1980 (in thousands)"


# The first table is the populations between 1950 and 1980
fifty_to_eighty = tables[0]
# Print the head of the table to get the first 5 rows
print(
    "* The first 5 rows of the \"Estimate between the years 1950 to 1980 (in thousands)\":\n\n",
    fifty_to_eighty.head()  
)
print("\n")

#---------------------------------------------------------------------------
# How many countries does the Wikipedia page list their populations?


# The second table is the populations between 1985 and 2015
eighty_five_to_fifteen = tables[1]
# The third table is the populations between 2020 and 2050
twenty_to_fifty = tables[2]

# Using len() and .unique() to find out the number of different countries and territories in each table
print(
    "* The number of countries (and dependent territories) that the page list their population is:",
    len(fifty_to_eighty["Country (or dependent territory)"].unique())
)

# Just checking if the numbers of countries are consistent among three tables. They are. 
# print(len(eighty_five_to_fifteen["Country (or dependent territory)"].unique()))
# print(len(twenty_to_fifty["Country (or dependent territory)"].unique()))

print("\n")

#---------------------------------------------------------------------------
# Create a DataFrame containing all the data from the population tables. I.e., the DataFrame contains the 
# popultation and average annual growth data from 1950-2050. You need to combine the three population tables.


# Using df1.merge(df2, on=column) to combine 2 dataframes at a time. Here I chain them to join 3 tables together. 
combined_table = fifty_to_eighty.merge(
    eighty_five_to_fifteen, on='Country (or dependent territory)'
    ).merge(
    twenty_to_fifty, on='Country (or dependent territory)'
)

# Print the first 5 rows of the combined table
print(
    "* The first 5 rows of the combined table:\n\n",
    combined_table.head()
)

print("\n")

#---------------------------------------------------------------------------
# For the top 10 most populous countries at 1950, plot their population data from 1950-2050 (in 5-year interval). 
# The plot should use years 1950-2050 as x axis, and population numbers as y axis. The following figure shows the expected result.


# Sort the values in combined_table by "1950" column in descending order
des_combined_table = combined_table.sort_values(by=['1950'], ascending=False)

# Put the first 10 countries into a new dataframe (opt out the first row since it is the world population)
top_five_in_1950 = des_combined_table.iloc[1:11, :]

# Reset index for top_five_in_1950 
top_five_in_1950.reset_index(drop=True, inplace=True)

# Just a quick rename since the original header is too long. Set 'inplace=True' to make changes to the original dataframe
top_five_in_1950.rename(columns={'Country (or dependent territory)':'Country'}, inplace=True)

# Make a new table with only population values. 
# I used start:stop:step slicing method to get values from 1955 to 2050. Then I insert 'Country' and '1950' column into the table. 
years = top_five_in_1950.iloc[:, 2::2]
years.insert(loc=0, column='1950', value=top_five_in_1950['1950'].values)
years.insert(loc=0, column='Country', value=top_five_in_1950.Country.values)

# Set 'Country' column as index. Set 'inplace=True' to make changes to the original dataframe
years.set_index('Country', inplace=True)
# Delete the word 'Country' above the index column (or it will go into the legend of the graph)
del years.index.name
# Transpose the dataframe
years = years.T

# Plotting
fig = plt.figure()
ax = years.plot(kind='line',style='o-', figsize=(12,7)) # line graph
ax.autoscale(tight=False) # Provide the whitespace at the start of each line (or they will start from the edge of y axis)
ax.set_xticks([i for i in range(0, len(years.index))]) # set up locations of ticks on x axis
ax.set_xticklabels(years.index, rotation=40) # set up labels for ticks on x axis

ax.set_xlabel("Year") # set label for x axis
ax.set_ylabel("Population in Thousands") # set label for y axis
plt.show()

#---------------------------------------------------------------------------
# For the same top 10 most populous contries at 1950, plot the average annual growth for each previous 
# five-year period starting from 1955 to 2050 in 5-year interval. The following figure shows the expected result.


growth = top_five_in_1950.iloc[:, 3::2]
growth.insert(loc=0, column='Country', value=top_five_in_1950.Country.values)

growth.set_index('Country', inplace=True)
del growth.index.name
growth = growth.T


fig = plt.figure()
ax = growth.plot(kind='line',style='o-', figsize=(12,7)) 
ax.autoscale(tight=False)
ax.set_xticks([i for i in range(0, len(growth.index))]) 
ax.set_xticklabels(years.index[1:], rotation=40) # years.index[1:] since we're not taking '1950' as a label
ax.set_xlabel("Year") 
ax.set_ylabel("Average Annual Growth in Previous 5 years (%)") 
plt.show()



```

```python
# Starting from 1901 to the latest year in the dataset, extract the birth countries of the laureates at each year. 
# Plot a bar chart showing the number of unique birth countries of the laureates over the years.


# Before extracting the 'year' from the 'prizes' column of the 'laureates' table, let's see if any laureates had 2
# Nobel Prizes (which means they have 2 'year' values in their row).

more_than_one_pos = []

years = []

# Iterate through the dictionaries inside data['laureates']
for i in range(0, len(data['laureates'])-1):
    result = data['laureates'][i] # Iterate through one dict at a time
    df2 = pd.DataFrame(result['prizes'], columns=['year']) # Make a one-column dataframe out of the 'year' key:value pair
   
    if len(df2.year.values) > 1: # We want to know the position of the laureates winning more than one year
        more_than_one_pos.append(i) # Append the position into the list
    
    for year in df2.year.values: # Store the years in the list
        years.append(year)

# The idea is to create a new 'Year' column for the 'laureates' dataframe. 
# Since each row can only contain one value for the 'year' column, we have to insert a duplicate row for each laureate
# who won more than one year.

# Create a dataframe list with 'df' in it
df_list = [df]

# Create duplicate rows in the form of new dataframes
for i in more_than_one_pos:
    df1 = df.loc[[i],:]
    df_list.append(df1) # Append new dataframes to df_list
    
# Concatenate all of the dataframes in the list to make a new dataframe with duplicate rows inserted
df_new = pd.concat(df_list).sort_index().reset_index(drop=True) 

# Create new column 'Year'
df_new['Year'] = years

# Use DataFrame.groupby() to extract the unique pair of 'bornCountry' and 'year' (omit duplicate birth countries in the same year)
born_and_year = df_new.groupby(['bornCountry', 'Year']).size().reset_index(name='count')

# Sort the dataframe on years in ascending order
born_and_year = born_and_year.sort_values(by=['Year'])

# Check the first 50 rows of the table
#print(born_and_year.head(50))

# Get the total count of unique birth countries in each year
year_count_df = born_and_year.groupby('Year').size().reset_index(name='Counts')

# Check the head and tail of year_count_df
#print(
#    year_count_df.head(),
#    year_count_df.tail() 
#)

# Plotting
tick_values = [i for i in range(0, len(year_count_df.Year.values)+5, 5)] # set xticks values
label_values = year_count_df.Year[::5] # set xticklabels values

fig = plt.figure()
ax = year_count_df.plot(
            kind='bar', 
            color='blue', 
            legend=False, 
            figsize=(12,7) 
)
ax.set_xticks(tick_values)
ax.set_xticklabels(label_values, rotation=40) 
ax.set_xlabel('Country', fontsize=13)
ax.set_ylabel('Number of Unique Countries', fontsize=13)
ax.set_title('Number of Countries the Laureates Were Born over the Years', fontsize=13)
plt.show()

```
