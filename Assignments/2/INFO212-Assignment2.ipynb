#### Question 2:
Load the data in the file 'inflammation-05.csv' to an Numpy array. Answer the following questions:
* Suppose the rows represent patients, and columns represent days. How many patients and how many days does the file contain?
* Extract the values for the last 10 patients' inflammation in last 10 days.
* For each patient, compute the average inflammation over all days. You should compute all patients' average inflammation values in single line. The result is an array. Plot all patients' averages and label the plot with "Average Inflammation of Patients". 
* For each day, compute the average inflammation over all patients. You should compute all the daily average inflammation values in a single line. The result is an array. Plot all daily averages and label the plot with "Average Inflammation of Days" 

````python
import numpy as np # Import Numpy
import matplotlib.pyplot as plt # Import matplotlib
%matplotlib inline 

# Use set_printoptions() with threshold=np.inf (infinity) to print out the array at full size
np.set_printoptions(threshold=np.inf) 

# Use loadtxt() from Numpy to load csv data into an array
inflam_05_arr = np.loadtxt(r"/Users/macbook/Desktop/INFO212/Assignment 2/datasets/inflammation-data/inflammation-05.csv", delimiter=',')

#--------------------------------------------------------------------
# Suppose the rows represent patients, and columns represent days. How many patients and how many days does the file contain?

num_patients, num_days = inflam_05_arr.shape # use .shape to get the dimension of the array
print("Number of patients:", num_patients, '\n'+ # print the answer
      "Number of days:", num_days)

#--------------------------------------------------------------------
# Extract the values for the last 10 patients' inflammation in last 10 days.

# Simple numpy array slicing
# [-10:, to get the last 10 rows
# ,-10:] to get the last 10 columns
print(inflam_05_arr[-10:,-10:])

#--------------------------------------------------------------------
# For each patient, compute the average inflammation over all days. You should compute all patients' average inflammation values 
# in single line. The result is an array. Plot all patients' averages and label the plot with "Average Inflammation of Patients".

# Use np.mean(data, axis=1) to get the average values across axis 1, which are rows
patients_avg = np.mean(inflam_05_arr, axis=1)

# Plotting
fig = plt.figure()
fig.suptitle('Average Inflammation of Patients', size=15) # Setting title
plt.plot(patients_avg) # Show the figure with the plot

#--------------------------------------------------------------------
# For each day, compute the average inflammation over all patients. You should compute all the daily average inflammation values 
# in a single line. The result is an array. Plot all daily averages and label the plot with "Average Inflammation of Days"

# Use np.mean(data, axis=0) to get the average values across axis 0, which are columns
daily_avg = np.mean(inflam_05_arr, axis=0)

# Plotting
fig = plt.figure()
fig.suptitle('Average Inflammation of Days', size=15) # Setting title
plt.plot(daily_avg) # Show the figure with the plot
````

#### Question 5:
Continue to work on the DataFrame of wine review loaded in Question 4. Answer the following questions:
* What countries are represented in the data set? i.e., list all the unique country names.
* What are the top-10 countries with the most wine reviews?
* What is the median of the price?
* Create a new column 'price_norm' by subtracting the median price from the price column. Use the DataFrame.apply method.
* Create a new column 'bargain' which contains the points-to-price ratio. Which wine(s) is/are the "best bargain", e.g., which wine(s) has/have the highest points-to-price ratio in the dataset?

````python
# What countries are represented in the data set? i.e., list all the unique country names.

print('Here are the countries represented in the data set:')

# Create a list of unique country names
country_list = list(df['country'].unique())

# Remove nan value from the list
country_list.remove(np.nan)

# Use print(*list, sep=', ') to print the elements without '\n' at the end of each print like when using for loop.
print(*country_list, sep=', ')

print('\n')

#--------------------------------------------------------------------
# What are the top-10 countries with the most wine reviews?

# Create an empty list
num_of_reviews = []

# For loop that iterates through the country_list
for country in country_list:
    country_df = df[df['country']==country] # Get a new dataframe that contains the reviews of 'country' only
    num_of_reviews.append(country_df.shape[0]) # Append number of reviews for each country into num_of_reviews

# Create a dictionary with 2 pairs of key:value
country_reviews_dict = {
    'Country': country_list,
    'Number of reviews': num_of_reviews
}

# Convert it to dataframe
country_reviews_df = pd.DataFrame(country_reviews_dict)

# Sort the dataframe by 'Number of reviews' column in descending order
des_country_reviews_df = country_reviews_df.sort_values(by=['Number of reviews'], ascending=False)

print('Top ten countries with most reviews:')
print(des_country_reviews_df.iloc[:10, :]) # Get the top-10

print('\n')

#--------------------------------------------------------------------
# What is the median of the price?

# Use df.median(axis=0, skipna=True) to get median values of numeric columns, and skip NaN values
print(df.median(axis=0, skipna=True)) 

# From the result, we have the median price of 25.0
print('The median of the price is 25.0.')

print('\n')

#--------------------------------------------------------------------
# Create a new column 'price_norm' by subtracting the median price from the price column. 
# Use the DataFrame.apply method.

# Use df.apply() to create a new column 'price_norm'
df['price_norm'] = df.apply(lambda row: (row.price - 25) , axis=1)

# See the 3 columns 'country', 'price', and 'price_norm'
print(df.loc[:, ['country', 'price', 'price_norm']])

print('\n')

#--------------------------------------------------------------------
# Create a new column 'bargain' which contains the points-to-price ratio. Which wine(s) is/are 
# the "best bargain", e.g., which wine(s) has/have the highest points-to-price ratio in the dataset?

# Create a Series that contains points-to-price ratio
pnt_to_prc = df['points']/df['price']

# Create a new column 'bargain'
df['bargain'] = pnt_to_prc

# See the 3 columns 'country', 'title', 'bargain'
bargain_df = df.loc[:, ['country', 'title', 'bargain']]
print('Let\'s see three columns "country", "title", and "bargain":')
print(bargain_df)
print('\n')

# See the dataframe sorted by 'bargain' column in descending order
sorted_bargain_df = bargain_df.sort_values(by=['bargain'], ascending=False)
print('Same dataframe but sorted by "bargain" column in descending order:')
print(sorted_bargain_df)

# See the top-5 of the dataframe
print('\n')
print('The top 5 "best bargain" wines:')
print(sorted_bargain_df.head())

print(df.price.mean())
````
